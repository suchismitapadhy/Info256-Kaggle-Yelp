### Imports

```python
import nltk,re
import pandas as pd
import numpy as w
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("yelp_data_official_training.csv",'r', delimiter="|")
df= df[["Category","Review Text"]]
df= df[df["Review Text"].notnull()]
random_index = np.random.permutation(df.index)
df = df.ix[random_index]
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
rows, columns = df.shape
train_size = round(rows*.8)
dev_size   = round(rows*.2)
```

```python
df_train = df.loc[:train_size]
df_dev = df.loc[train_size:dev_size+train_size].reset_index(drop=True)
df_train.shape
```

```python
vec = CountVectorizer(ngram_range=(1,2), token_pattern=r'\b\w+\b', min_df=5, max_features=2000, stop_words="english")  
tokenizer = vec.build_tokenizer()
```

```python
arr_train_feature_sparse = vec.fit_transform(df_train["Review Text"])
arr_train_feature = arr_train_feature_sparse.toarray()
```

```python
arr_dev_feature_sparse = vec.fit_transform(df_dev["Review Text"])
arr_dev_feature = arr_dev_feature_sparse.toarray()
```

```python
nb = MultinomialNB()
nb_model = nb.fit(arr_train_feature, df_train.Category)
nb_predictions = nb_model.predict(arr_dev_feature)
accuracy_score(df_dev.Category, nb_predictions)
```

```python
logreg = LogisticRegression()
logreg_model = logreg.fit(arr_train_feature, df_train.Category)
logreg_predictions = logreg_model.predict(arr_dev_feature)
accuracy_score(df_dev.Category, logreg_predictions)
```

```python
def tokenize(text):
    pattern= r'''(?x) (?:[A-Z]\.)+|(?:\w+(?:[-']\w+)*)'''
    tokens = (nltk.regexp_tokenize(text,pattern))
    wnlemmatizer = nltk.WordNetLemmatizer()
    wnlemma_tokens = [wnlemmatizer.lemmatize(word.lower()) for word in tokens if word.isalpha()]
    stemmed_tokens = nltk.PorterStemmer.
    return wnlemma_tokens
```

```python
vec2 = TfidfVectorizer(tokenizer=tokenize, max_features=2000, stop_words="english",ngram_range=(1,2))
tokenizer2 = vec2.build_tokenizer()
```

```python
tfidf_train_feature_sparse = vec2.fit_transform(df_train["Review Text"])
tfidf_train_feature = tfidf_train_feature_sparse.toarray()

tfidf_dev_feature_sparse = vec2.fit_transform(df_dev["Review Text"])
tfidf_dev_feature = tfidf_dev_feature_sparse.toarray()
```

```python
nb = MultinomialNB()
nb_model = nb.fit(tfidf_train_feature, df_train.Category)
nb_predictions = nb_model.predict(tfidf_dev_feature)
accuracy_score(df_dev.Category, nb_predictions)
```

```python
logreg = LogisticRegression()
logreg_model = logreg.fit(tfidf_train_feature, df_train.Category)
logreg_predictions = logreg_model.predict(tfidf_dev_feature)
accuracy_score(df_dev.Category, logreg_predictions)
```
